```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Load the data from a local file (ensure the file is in the same directory as your Jupyter Notebook)
df = pd.read_csv("honeyproduction.csv")

# Display the first few rows of the dataset to inspect its structure
df.head()

# Calculate the mean honey production per year
prod_per_year = df.groupby('year')['totalprod'].mean().reset_index()

# Display the resulting DataFrame
prod_per_year

# Extract the features (years) and the target variable (total honey production)
X = prod_per_year[['year']]
y = prod_per_year['totalprod']

# Create a scatter plot of the data
plt.figure(figsize=(10, 6))
plt.scatter(X, y)
plt.xlabel('Year')
plt.ylabel('Total Honey Production')
plt.title('Honey Production Over the Years')
plt.grid(True)
plt.show()

# Create a linear regression model and fit it to the data
regr = LinearRegression()
regr.fit(X, y)

# Print the coefficient and intercept of the linear regression model
print("Coefficient (Slope):", regr.coef_[0])
print("Intercept:", regr.intercept_)

# Make predictions using the trained model
y_predict = regr.predict(X)

# Plot the regression line
plt.figure(figsize=(10, 6))
plt.plot(X, y_predict, color='red')
plt.xlabel('Year')
plt.ylabel('Total Honey Production')
plt.title('Linear Regression Fit')
plt.grid(True)
plt.show()

# Generate future years for prediction
X_future = np.array(range(2013, 2050)).reshape(-1, 1)

# Predict the total honey production for future years
future_predict = regr.predict(X_future)

# Plot the predicted values for future years
plt.figure(figsize=(10, 6))
plt.plot(X_future, future_predict, color='green')
plt.xlabel('Year')
plt.ylabel('Predicted Total Honey Production')
plt.title('Future Honey Production Prediction')
plt.grid(True)
plt.show()
```
